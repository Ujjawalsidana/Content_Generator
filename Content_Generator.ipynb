```python
from transformers import GPT2LMHeadModel,GPT2Tokenizer
```

```python
tokenize=GPT2Tokenizer.from_pretrained("gpt2")
model=GPT2LMHeadModel.from_pretrained("gpt2")
```

```python
text="What is Blockchain Development?"
encoded_input = tokenize.encode(text,return_tensors='pt')
```

```python
encoded_input
```

```python
tokenize.decode(encoded_input[0][0])
```

```python
outp=model.generate(encoded_input,max_length=500,num_beams=5,no_repeat_ngram_size=2,early_stopping=True)
```
